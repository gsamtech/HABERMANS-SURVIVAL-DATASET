NOTE: 
So from above PDF we can say the patients survival status, but we cannot exactly say what percentage of patient will actually short survive or long survive. To know that we have another distribution that is CDF.

```python
CDF of habarman_survival
```

```python
counts, bin_edges = np.histogram(habarman_survival['Axillary_node'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)


plt.show();
```

OBSERVATION:
The above CDF shows the percentage of long survival which is 85% if the number of axillary nodes is < 5
while 100% for short survival if the axillarynode is >40

```python
#Mean, Variance, Std-deviation,  
print("Means:")
print(np.mean(habarman_survival["Axillary_node"]))
#Mean with an outlier.
print(np.mean(np.append(habarman_survival["Axillary_node"],50)));
print(np.mean(habarman_nonsurvival["Axillary_node"]))


print("\nStd-dev:");
print(np.std(habarman_survival["Axillary_node"]))
print(np.std(habarman_nonsurvival["Axillary_node"]))


```

OBSERVATION:
You can observe that for Long survive mean is 2.79 and including outlier it is 3 that is almost same, but the mean of Short survive is 7.4 which is comparatively much higher than Long survive. So the probability for short survive is more in data set.
If you observe the standard deviation Long survive has standard deviation of only 2.79 and Short survive has 7.45, means the spread of data for short survive is more.

```python
print("\nMedians:")
print(np.median(habarman_survival["Axillary_node"]))
#Median with an outlier
print(np.median(np.append(habarman_survival["Axillary_node"],50)));
print(np.median(habarman_nonsurvival["Axillary_node"]))



print("\nQuantiles:")
print(np.percentile(habarman_survival["Axillary_node"],np.arange(0, 100, 25)))
print(np.percentile(habarman_nonsurvival["Axillary_node"],np.arange(0, 100, 25)))


print("\n90th Percentiles:")
print(np.percentile(habarman_survival["Axillary_node"],90))
print(np.percentile(habarman_nonsurvival["Axillary_node"],90))


from statsmodels import robust
print ("\nMedian Absolute Deviation")
print(robust.mad(habarman_survival["Axillary_node"]))
print(robust.mad(habarman_nonsurvival["Axillary_node"]))

```

From above observation it is clear that average axillary nodes in long survival is 0 and for short survival it is 4. ie, Patients who have average 4 auxillary nodes have short survival status.

Quantiles shows that nearly 50th% of axillary nodes are 0 in long survival and 75th% of patients have nodes less than 3 that is 25% patients are having nodes more than 3.
At 90th% there if nodes detected is >8 then it has long survival status and if nodes are >20 then patients will have short survival status
